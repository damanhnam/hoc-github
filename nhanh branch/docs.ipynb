### 3. **Xử lý branch trong Git**

*Bảng các lệnh thao tác với branchs -*  ***git branching*** 
| Lệnh                         | Mô tả                                                                |
|------------------------------|----------------------------------------------------------------------|
| `git branch`                 | Liệt kê danh sách branches               |
| `git checkout -b <tên nhánh>` | Tạo nhánh mới , không có -b tức là chuyến đến nhánh đã có|
| `git branch -D  <Nhánh A>`                 | Xoá nhánh A khỏi local                |
| `git merge <Nhánh A>`                 | Ghép nhánh A vào nhánh hiện tại                |
| `git merge --abort`                 | Huỷ việc ghép nhánh đang xảy ra                 |
