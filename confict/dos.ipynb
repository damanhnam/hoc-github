### 4. **Xử lý conflict Git**

Conflict xảy ra khi việc thực hiện ghép giứa hai nhánh đang có cùng sự thay đổi trên 1 file. Hoặc 1 file đang có sự xung đột giữa hai phiên bản local và remote  

| Lệnh                     | Mô tả                                                                 |
|---------------------------------------|----------------------------------------------------------------------|
| `git pull`                            | Kéo code từ remote về — dễ xảy ra conflict nếu có thay đổi trùng nhau |
| `git merge <branch>`                  | Gộp nhánh — có thể gây ra conflict nếu có sửa đổi cùng vị trí        |
| **Khi xảy ra conflict:**              |                                                                      |
| `git merge --continue`         | Tiếp tục sau khi đã xử lý conflict trong quá trình merge            |
| `git merge --abort`                  | Hủy quá trình merge và quay về trạng thái trước đó                  |
| `git status`                          | Kiểm tra các file còn đang bị conflict                              |

Quá trình thực hiện xử lý conflict:  
- Sau khi thực hiện `git merge <branch>`, nếu có sự xung đọt, việc merge sẽ dừng lại và chỉ thực hiện staged các files không có conflict. Các file trong trang thái conflict (xung đột) sẽ được đánh dấu và chuyển vào giao diện merge edditor
- Cấc đoạn mã có sự xung đột sẽ được đánh dấu cần thực hiện chỉnh sửa để khắc phục conflict
- Sau khi xử lý xung đột ở các file, thực hiện staged các file để đánh dấu rằng file đã được khắc phục conflict, tiếp tục đến khi không còn file nào bị đánh dấu xung đột.
