### **1. Các lệnh cơ bản của Linux**  
| STT | Chức năng                | Lệnh                                  | Ghi chú                             |
| --- | ------------------------ | ------------------------------------- | ----------------------------------- |
| 1   | Tạo thư mục              | `mkdir <tên_thư_mục>`                 | Tạo thư mục mới trong thư mục hiện tại|
| 2   | Tạo file                 | `touch <tên_file>`                    | Tạo file rỗng                       |
| 3   | Sửa file                 | `nano <tên_file>`                     | Mở trình chỉnh sửa văn bản          |
| 4   | Liệt kê file/thư mục     | `ls`                                  | Liệt kê nội dung thư mục hiện tại   |
|     |                          | `ls -l`                               | Liệt kê chi tiết từng file          |
|     |                          | `ls -a`                               | Hiện cả file ẩn                     |
|     |                          | `ls -lh`                              | Hiển thị dung lượng dễ đọc          |
| 5   | Xem nội dung file        | `cat <tên_file>`                      | Xem toàn bộ nội dung                |
|     |                          | `head <tên_file>` / `tail <tên_file>` | Xem dòng đầu/cuối của file          |
| 6   | Xóa file/thư mục         | `rm <tên_file>`                       | Xóa file                            |
|     |                          | `rm -r <tên_thư_mục>`                 | Xóa thư mục và nội dung             |
|     |                          | `rm -rf <tên_thư_mục>`                | Xóa, không hỏi xác nhận             |
| 7   | Đổi tên/Di chuyển        | `mv <nguồn> <đích>`                   | Di chuyển hoặc đổi tên file/thư mục |
| 8   | Sao chép file/thư mục    | `cp <nguồn> <đích>`                   | Sao chép file                       |
|     |                          | `cp -r <thư_mục_nguồn> <đích>`        | Sao chép thư mục                    |
| 9   | Chuyển thư mục           | `cd <thư_mục>`                        | Vào thư mục cụ thể                  |
|     |                          | `cd ..`                               | Quay về thư mục cha                 |
|     |                          | `cd ~`                                | Về thư mục home                     |
| 10  | Kiểm tra vị trí hiện tại | `pwd`                                 | In ra đường dẫn thư mục hiện tại    |   

*Thực hành trên WSL (Window System Linux) - thư mục: ./linux-commands*
