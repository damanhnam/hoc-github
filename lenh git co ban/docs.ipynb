```python
### **2. Các lệnh cơ bản của Git**
#### Git:
Git là hệ thống quả lý phiên bản phân tán, giúp theo dõi và quản lý sự thay đối của mã nguồn trong quá trình phát triển phân mềm.

*Bảng các lệnh cơ bản trong git*

| Lệnh                         | Mô tả                                                                |
|------------------------------|----------------------------------------------------------------------|
| `git init`                   | Khởi tạo một repository Git mới trong thư mục hiện tại               |
| `git clone <url>`            | Sao chép một repository từ xa về máy local                           |
| `git status`                 | Hiển thị trạng thái các tệp đã thay đổi và chưa được commit          |
| `git add <file>`             | Thêm tệp vào khu vực staging để chuẩn bị commit                      |
| `git add .`                  | Thêm tất cả các thay đổi vào staging                                 |
| `git commit -m "message"`    | Lưu thay đổi vào lịch sử với một thông điệp                          |
| `git log`                    | Xem lịch sử các commit                                               |
| `git diff`                   | Hiển thị sự khác biệt giữa các thay đổi                              |
| `git branch`                 | Liệt kê các nhánh hoặc tạo nhánh mới nếu dùng `git branch <name>`    |
| `git checkout <branch>`      | Chuyển sang nhánh khác                                               |
| `git merge <branch>`         | Gộp nhánh vào nhánh hiện tại                                         |
| `git pull`                   | Tải và hợp nhất thay đổi từ remote                                   |
| `git push`                   | Đẩy các commit lên remote                                            |
| `git remote -v`              | Hiển thị thông tin các remote repository                             |
| `git reset --soft HEAD^ `    | Bóc commit cuối cùng, giữ lại các thay đối trong staged              |
| `git rm <file>`              | Xóa file và stage thay đổi                                                                |
| `git stash pop`              | Khôi phục lại các thay đổi đã stash                                  |


 


```
